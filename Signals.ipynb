```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Define time range
t = np.linspace(0, 1, 1000)
# Define frequency
f = 5

analog_signal = np.sin(2*np.pi*f*t)

# Plotting
plt.figure()
plt.plot(t, analog_signal)
plt.title("Analog Signal (Sine Wave)") 
plt.xlabel('Time (S)') 
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Define time index n
n = np.arange(0, 1, 0.02)
# Define the frequency
f = 5
# Define the digital signal as sampled sine wave
digital_signal = np.sin(2*np.pi*f*n)

# Plotting
plt.figure()
plt.stem(n, digital_signal, basefmt=" ", use_line_collection=True)
plt.title('Digital Signal (Sampled Sine Wave)')
plt.xlabel('Discrete Time (Samples)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

```

```python
# Time shift
# Define time 
t = np.linspace(-1, 1, 1000)
# Define frequency
f = 2
# Origial signal
x = np.sin(2 * np.pi * f * t)

# Shift original with 0.5 seconds
shift = 0.1
shifted_x = np.sin(2 * np.pi * f * (t - shift))

# Plotting
plt.figure()
plt.plot(t, x, label='Original Signal')
plt.plot(t, shifted_x, label=f'Shifted Signal (t - {shift})')
plt.title('Time shift')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
# Time Scaling
# Define time 
t = np.linspace(-1, 1, 1000)
# Define the frequency
f = 2
# Original signal
x = np.sin(2 * np.pi * f * t)

# Time scaling with factor of 2
scale = 2
scaled_x = np.sin(2 * np.pi * f * (t * scale))

# Plotting
plt.figure()
plt.plot(t, x, label='X')
plt.plot(t, scaled_x, label=f'Scaled (t * {scale})')
plt.title('Time Scaling')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
# Time folding
# Define time
t = np.linspace(-1, 1, 1000)
# Define Frequency
f = 0.5
# Original Signal
x = np.sin(2 * np.pi * f * t)

# Time folding
folded_x = np.sin(2 * np.pi * f * (-t))

# Plotting
plt.figure()
plt.plot(t, x, label='Original Signal')
plt.plot(t, folded_x, label='Folded Signal')
plt.title('Time Folding')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```
