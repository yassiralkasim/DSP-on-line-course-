```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 1000)
f = 5
x = np.sin(2 * np.pi * f * t)

fs = 50

n_samples = np.arange(0, 1, 1/fs)
x_n = np.sin(2 * np.pi * f * (n_samples))

# plotting
plt.figure()
plt.plot(t, x, label='Analog Signal')
plt.stem(n_samples, x_n, basefmt=" ", linefmt="r", markerfmt="ro", label="Sampled Signal", use_line_collection=True)
plt.title('Sampling Theory Demonstration')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
t = np.linspace(0, 1, 1000)
f = 1
x = np.sin(2 * np.pi * f * t)
quan_levels = 8
q_x = np.round(x * (quan_levels / 2)) / (quan_levels / 2)

# plotting
plt.figure()
plt.plot(t, x, label='Analog Signal')
plt.step(t, q_x, label='Quantized Signal')
plt.title('Qauntization of Analog Signal')
plt.xlabel('Time (S)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```
